# Clean price, discount and rating_count columns

```python
df['discounted_price'] = df['discounted_price'].astype(str)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
print(df['discounted_price'])
```

```python
df['actual_price'] = df['actual_price'].astype(str)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
print(df['actual_price'])
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(str)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
print(df['discount_percentage'])
```

```python
df['rating_count'] = df['rating_count'].astype(str)
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
print(df['rating_count'])
```

# Handle missing values

```python
print(df['rating_count'].fillna(df['rating_count'].median(), inplace=True))
```

# Price difference

```python
print(df['price_difference'] == df['actual_price'].fillna(0) - df['discounted_price'].fillna(0))
```

# Average rating

```python
print("Average Product Rating:", df['rating'].mean())
```

# Top 5 categories (split and explode category string first)

```python
df['main_category'] = df['category'].str.split('|').str[0]
top_categories = df['main_category'].value_counts().head(5)
top_categories.plot(kind='bar', title='Top 5 Product Categories')
plt.xlabel("Category")
plt.ylabel("Product Count")
plt.tight_layout()
plt.show()
```

# Top 10 rated products

```python
top_rated = df.sort_values(by='rating', ascending=False)[['product_name', 'discounted_price', 'rating']].head(10)
print("Top 10 Rated Products:\n", top_rated)
```

# Histogram of discount percentages

```python
plt.hist(df['discount_percentage'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title("Discount Percentage Distribution")
plt.xlabel("Discount %")
plt.ylabel("Number of Products")
plt.tight_layout()
plt.show()
```

# Average discount by category

```python
avg_discount_by_cat = df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)
print("Average Discount by Category:\n", avg_discount_by_cat)
```

# Line chart of rating count

```python
plt.plot(df['rating_count'].head(100).values)
plt.title("Rating Count for First 100 Products")
plt.xlabel("Product Index")
plt.ylabel("Rating Count")
plt.tight_layout()
plt.show()
```

# Highest average rating category

```python
avg_rating_by_cat = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)
print("Category with Highest Average Rating:\n", avg_rating_by_cat)
```

# Price vs Rating scatter plot

```python
plt.scatter(df['discounted_price'], df['rating'], alpha=0.5)
plt.title("Price vs Rating")
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```

# Most reviewed products

```python
top_reviewed = df.sort_values(by='rating_count', ascending=False)[['product_name', 'rating_count']].head()
print("Top 5 Most Reviewed Products:\n", top_reviewed)
```

# Revenue estimation

```python
df['estimated_revenue'] = df['discounted_price'] * df['rating_count']
top_revenue = df.sort_values(by='estimated_revenue', ascending=False)[['product_name', 'estimated_revenue']].head()
print("Top 5 Estimated Revenue Products:\n", top_revenue)
```

# Subplots for rating count and rating

```python
top5 = df.sort_values(by='rating_count', ascending=False).head(5)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(top5['product_name'], top5['rating_count'], color='orange')
plt.title("Top 5 Products by Rating Count")
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
plt.plot(top5['product_name'], top5['rating'], marker='o', color='green')
plt.title("Rating of Top 5 Reviewed Products")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
